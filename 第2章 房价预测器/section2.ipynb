```python
### PyTorch实例：预测房价
```

```python
# 准备数据
x = torch.linspace(0, 100, 100).type(torch.FloatTensor)
rand = torch.randn(100) * 10
y = x + rand
x_train = x[: -10]
y_train = y[: -10]
x_test = x[-10 :]
y_test = y[-10 :]
import matplotlib.pyplot as plt
plt.plot(x_train.data.numpy(), y_train.data.numpy(), 'o')
plt.xlabel('X')
plt.ylabel('Y')
  
```

```python
#设计模型与训练
a = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)
learning_rate = 0.0001
for i in range(1000):
    predictions = a.expand_as(x_train) * x_train + b.expand_as(x_train)
    loss = torch.mean((predictions - y_train) ** 2)
    print('loss=', loss)
    loss.backward()
    a.data.add_(-learning_rate * a.grad.data)
    b.data.add_(-learning_rate * b.grad.data)
    a.grad.data.zero_()
    b.grad.data.zero_()
#作图
plt.plot(x_train.data.numpy(), y_train.data.numpy(), 'o')
plt.xlabel('X')
plt.ylabel('Y')
plt.plot(x_train.data.numpy(), a.data.numpy() * x_train.data.numpy() + b.data.numpy())
plt.show()
```

```python
# 后十年的预测
x_data = x_test.data.numpy()
y_data = y_test.data.numpy()
predictions = a.data.numpy() * x_data + b.data.numpy()
plt.plot(x_data, y_data, 'o')
plt.plot(x_data, predictions)
plt.show()
```
